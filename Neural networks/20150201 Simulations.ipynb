```python
#import matplotlib.pyplot as plt

import pandas as pd
#pd.options.display.mpl_style = 'default'


import numpy as np
explanatory_vars = ["x"+str(i) for i in range(1,6)]
df_simple = pd.DataFrame((np.random.rand(10000, len(explanatory_vars))-0.5)*20, columns=explanatory_vars)
df_nonlinear = pd.DataFrame((np.random.rand(10000,len(explanatory_vars))-0.5)*5, columns=explanatory_vars)
```

```python
import scipy.special as special

#This is the real data generating process
df_simple["x"] = 3 + 0.7*df_simple["x1"] + 1.5*df_simple["x2"]

#Calculate the probability from the logit function
df_simple["yprob"] = special.expit(df_simple["x"])

#Calculate some uniform draws from error
df_simple["e"] = np.random.rand(df_simple.shape[0])

#Compute the observed data - i.e. 1s and 0s - according to generated probabilities
df_simple["y"] = 1
df_simple["y"]= df_simple["y"].where(df_simple["yprob"]>df_simple["e"],0)
df_simple.head()
```

```python
import scipy.special as special

#This is the real data generating process
df_nonlinear["x"] =3*df_nonlinear["x1"] + 1*df_nonlinear["x2"]  + 3*df_nonlinear["x1"]*df_nonlinear["x2"] -0.2*df_nonlinear["x1"]**3

#Calculate the probability from the logit function
df_nonlinear["yprob"] = special.expit(df_nonlinear["x"])

#Calculate some uniform draws from error
df_nonlinear["e"] = np.random.rand(df_nonlinear.shape[0])

#Compute the observed data - i.e. 1s and 0s - according to generated probabilities
df_nonlinear["y"] = 1
df_nonlinear["y"]= df_nonlinear["y"].where(df_nonlinear["yprob"]>df_nonlinear["e"],0)
```

```python
import scipy
from sklearn import svm
```

```python
clf = svm.SVC(kernel = 'rbf',  gamma=0.7, C=3 )
X = np.array(df_nonlinear[["x1","x2"]])
Y = np.array(df_nonlinear["y"])
clf.fit(X,Y)
```

```python

fig = plt.figure()


plt.scatter(df_nonlinear["x1"], df_nonlinear["x2"],c=df_nonlinear['y'], edgecolors='none', alpha=0.1,s=5)
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))


# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, cmap=plt.cm.Paired)
```
